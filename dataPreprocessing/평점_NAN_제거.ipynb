```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2) # KNN 활용해서 인접한 2개의 이웃을 반영해 NaN값 대치
```

```python
import numpy as np
import pandas as pd
```

```python
pd.DataFrame(rate)
```

```python
평론가_df = rate[['평론가 평점']] # fit 위해 2차원으로 만들어 줌
```

```python
네티즌_df = rate[['네티즌 평점']]
```

```python
평론가_filled = imputer.fit_transform(평론가_df)
```

```python
네티즌_filled = imputer.fit_transform(네티즌_df)
```

```python
pd.DataFrame(평론가_filled)
```

```python
pd.DataFrame(네티즌_filled)
```

```python
rate.drop(['평론가 평점'], axis=1)
rate['평론가 평점'] = pd.DataFrame(평론가_filled)
```

```python
rate.drop(['네티즌 평점'], axis=1)
rate['네티즌 평점'] = pd.DataFrame(네티즌_filled)
```

```python
rate = rate.round(2)
```
